# data augmemntation of covid images 

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
datagen=ImageDataGenerator(
rotation_range=40,width_shift_range=0.2,
height_shift_range=0.2,
shear_range=0.2,
zoom_range=0.2,
horizontal_flip=True,
fill_mode='nearest')
```

```python
files=glob.glob('COVID2_CT\*.*')
for file in files[:500]:
    img=load_img(file)
    x=img_to_array(img)
    x=x.reshape((1,)+x.shape)
    j=0
    for batch in datagen.flow(x,batch_size=1,save_to_dir='covidnew',save_prefix='cov',save_format='jpeg'):
        j+=1
        if j>2:
            break
            
```

#### number of augmented images in covide Ct scan

```python

files=glob.glob('covidnew\*.*')
count=0
for file in files:
    count+=1
print(count)

```

### augumenting images in NORMAL CT

```python
files=glob.glob('NORMAL_CT\*.*')
for file in files[:500]:
    img=load_img(file)
    x=img_to_array(img)
    x=x.reshape((1,)+x.shape)
    j=0
    for batch in datagen.flow(x,batch_size=1,save_to_dir='normalnew',save_prefix='norm',save_format='jpeg'):
        j+=1
        if j>2:
            break
            
```
