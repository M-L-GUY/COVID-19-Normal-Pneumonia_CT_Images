```python
import glob
import cv2
import numpy
import random
```

```python
files=glob.glob('COVID2_CT\*.*')
```

```python
data=[]
for file in files:
    new_img=cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    new_img=cv2.resize(new_img,(200,200))
    data.append([new_img,0])
```

```python
len(data)
```

```python
import matplotlib.pyplot as plt
plt.imshow(data[55][0])
plt.xlabel(data[55][1])
```

```python
files=glob.glob('NORMAL_CT/*.*')
for files in files:
    new_img=cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    new_img=cv2.resize(new_img,(200,200))
    data.append([new_img,1])
```

```python
len(data)
```

```python
files=glob.glob('pneumonia_CT/*.*')
for files in files:
    new_img=cv2.imread(file,cv2.IMREAD_GRAYSCALE)
    new_img=cv2.resize(new_img,(200,200))
    data.append([new_img,2])
```

```python
print(len(data))
random.shuffle(data)
```

```python
img_data=[]
img_label=[]
for DATA,LABEL in data:
    img_data.append(DATA)
    img_label.append(LABEL)
```

```python
print(len(img_data))
```

```python
print(len(img_label))
```

```python
import matplotlib.pyplot as plt
CATEGORIES=['COVID','NORMAL','PNEUMONIA']
plt.figure(figsize=(20,20))
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.imshow(img_data[i],cmap='gray')
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(CATEGORIES[img_label[i]])
```
